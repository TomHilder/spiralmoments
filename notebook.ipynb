```python
from velocity_field import VelocityField
from plots          import spiral_v_contours_plot
```

```python
import matplotlib.pylab as plt
import numpy            as np
```

```python
# create velocity field objects for wakeflow, phantom and observations
wf0 = VelocityField(type="wakeflow", name="3.0Mj")
wfT = VelocityField(type="wakeflow", name="3.0Mj_total")
ph = VelocityField(type="phantom", name="hd163")
ob = VelocityField(type="observations", name="HD_163296_CO_220GHz.0.2arcsec.JvMcorr.image_v0.fits")
```

```python
# rotate wf0, wfT and ph
wf0.rotate(PA=45, planet_az=45, inc=-225)
wfT.rotate(PA=45, planet_az=45, inc=-225)
ph.rotate(PA=45, planet_az=45, inc=-225)
```

```python
# concert to angular coordinates
wf0.convert_to_angular_coordinates()
wfT.convert_to_angular_coordinates()
ph.convert_to_angular_coordinates()
```

```python
contour_lvls = np.linspace(-2000,2000,40)
spiral_v_contours_plot(spiral=wf0, contours=wfT, contour_levels=contour_lvls)
```

```python
contour_lvls = np.linspace(-2000,2000,40)
spiral_v_contours_plot(spiral=wf0, contours=ph, contour_levels=contour_lvls)
```

```python
contour_lvls = np.linspace(-2000,2000,40)
spiral_v_contours_plot(spiral=wf0, contours=ob, contour_levels=contour_lvls)
```
